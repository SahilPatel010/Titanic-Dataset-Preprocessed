```python
df['Age'].unique().sum()
```

```python
df['Age'] = df['Age'].fillna(df.groupby('Pclass1')['Age'].transform('mean'))
df['Age'] = df['Age'].fillna(df.groupby('Pclass2')['Age'].transform('mean'))
df['Age'] = df['Age'].fillna(df.groupby('Pclass3')['Age'].transform('mean'))
```

```python
import numpy as np

# This will round every age to the nearest whole number and then convert it to an integer.
df['Age']=df['Age'].round().astype(int)

# Check your work. Now infants will be 1 instead of 0.
df.head()
```
